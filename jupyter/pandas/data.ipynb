```python
#写入单个文件

import pandas as pd #导入pandas
AlCu='input_Al_Cu_system.data'

#在终端运行 awk '/begin/ {print FNR}' input_Al_Cu_system.data >begin 将begin行号保存到begin文件
#在终端运行 awk '/end/ {print FNR}' input_Al_Cu_system.data >end 将end行号保存到end文件
with open('begin') as begin: #读入begin字符行数
    b=begin.readlines()
with open('end') as end: #读入end字符行数
    e=end.readlines()
    
i=0
POSCAR=f'{i}' #将文件名赋予POSCAR

with open(POSCAR,'w') as poscar:
    poscar.write('POSCAR\n')
    poscar.write('1.0\n')

lattice=pd.read_table(AlCu, header=None, names=None, sep='\s+', 
                         skiprows=int(b[i])+1, nrows=3, engine='python') #读取的行包含晶胞常数
del lattice[0] #删除第一列的atom
lattice.to_csv(POSCAR,index=False, header=False, sep=' ', mode='a', quoting=3) #写入晶胞常数

atom=pd.read_table(AlCu, header=None, names=None, sep='\s+', 
                         skiprows=int(b[i])+4, nrows=int(e[i])-int(b[i])-7, engine='python') ##读取的行包含原子位置
atom = atom.drop(labels=[0,5,6,7,8,9], axis=1) #删除原子多余列位置
atom.sort_values(by=4, ascending=True, inplace=True) #列表按照Al-Cu顺序排序
data=atom[4].tolist() #将Series数据转换成列表
al=data.count('Al') #列表中Al原子计数
cu=data.count('Cu') #列表中Cu原子计数

with open(POSCAR,'a') as poscar:
    if cu==0: #如果Cu原子不存在
        poscar.write('Al\n')
        poscar.write(f"{al}\n")
    elif al==0: #如果Al原子不存在
        poscar.write('Cu\n')
        poscar.write(f"{cu}\n")
    else: #如果Al和Cu原子同时存在
        poscar.write('Al Cu\n')
        poscar.write(f"{al} {cu}\n")
    poscar.write('Cartesian\n')
atom.to_csv(POSCAR, index=False, header=False, sep=' ', mode='a',quoting=3)
#写入各个原子位置；quoting=3避免写入数据双引号    
```

```python
#循环写入5267个文件

import pandas as pd #导入pandas
AlCu='input_Al_Cu_system.data'

#终端中运行 awk '/begin/ {print FNR}' input_Al_Cu_system.data >begin 将begin行号保存到begin文件
#终端中运行 awk '/end/ {print FNR}' input_Al_Cu_system.data >end 将end行号保存到end文件
with open('begin') as begin: #读入begin字符行数
    b=begin.readlines()
with open('end') as end: #读入end字符行数
    e=end.readlines()

i=0
while i<=5266:

    POSCAR=f'{i}' #将文件名赋予POSCAR
    with open(POSCAR,'w') as poscar:
        poscar.write('POSCAR\n')
        poscar.write('1.0\n')
    lattice=pd.read_table(AlCu, header=None, names=None, sep='\s+', 
                         skiprows=int(b[i])+1, nrows=3, engine='python') #读取的行包含晶胞常数
    del lattice[0] #删除第一列的atom
    lattice.to_csv(POSCAR,index=False,header=False,sep=' ',mode='a',quoting=3) #写入晶胞常数
    

    atom=pd.read_table(AlCu, header=None, names=None, sep='\s+', 
                         skiprows=int(b[i])+4, nrows=int(e[i])-int(b[i])-7, engine='python') #读取的行包含原子位置
    atom = atom.drop(labels=[0,5,6,7,8,9], axis=1) #删除原子多余列位置
    atom.sort_values(by=4, ascending=True, inplace=True) #列表按照Al-Cu顺序排序
    data=atom[4].tolist() #将Series数据转换成列表
    al=data.count('Al') #列表中Al原子计数
    cu=data.count('Cu') #列表中Cu原子计数

    with open(POSCAR,'a') as poscar:
        if cu==0: #如果Cu原子不存在
            poscar.write('Al\n')
            poscar.write(f"{al}\n")
        elif al==0: #如果Al原子不存在
            poscar.write('Cu\n')
            poscar.write(f"{cu}\n")
        else: #如果Al和Cu原子同时存在
            poscar.write('Al Cu\n')
            poscar.write(f"{al} {cu}\n")
        poscar.write('Cartesian\n')
    atom.to_csv(POSCAR,index=False,header=False,sep=' ',mode='a',quoting=3) 
    #写入各个原子位置；quoting=3避免写入数据双引号 
    i = i+1 
```
