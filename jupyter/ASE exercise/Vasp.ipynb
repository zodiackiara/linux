```python
from ase.calculators.vasp import Vasp
calc = Vasp(directory='./Pt') 
from ase.build import bulk
atoms = bulk('Pt', 'fcc', a=3.5, cubic=True)
atoms.calc = calc

print(atoms.get_potential_energy())
```

```python
from ase.calculators.vasp import Vasp
calc = Vasp(directory='./diamond',algo = 'fast',encut = 600,sigma=0.01,ediff=1E-5,nelm=100,ediffg=-0.01,nsw=100,
            prec='accurate',lwave=0,lreal='auto',lcharg=0,potim=0.1,
            gamma=1,kpts=(7,7,7),ispin=2,magmom=(0.6,0.6,0.6,0.6,0.6,0.6,0.6,0.6))
from ase.io import read
atoms=read('diamond.poscar')
atoms.calc = calc
atoms.get_potential_energy()
#kpts和magmom都需要输入列表，列表长度根据实际情况确定
#尝试isym=2,报错
#目前没有报错
```
