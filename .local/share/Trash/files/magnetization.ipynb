```python
from ase.atoms import Atoms
from ase.io import read
import pandas as pd

t = len(open('OUTCAR1.txt','rt').readlines())
pf=pd.read_csv('OUTCAR1.txt',skiprows=range(0,t-150),nrows=120,sep='\s+',header=None)
df=pf.drop(columns=0)
df.rename(columns={1:'s',2:'p',3:'d',4:'total'},inplace=True)

c=read('CONTCAR1.txt')
d=c.get_all_distances()
e=d[59]
e[59]=20
i,t=0,0
while i<119:
    if e[i]<3.6:
        t=t+1
        f=c[i].index
        print(f"The nearest atom is {c[i]}") 
        print(f"The magnetization is \n{df.loc[f]}\n")
    i=i+1
```
